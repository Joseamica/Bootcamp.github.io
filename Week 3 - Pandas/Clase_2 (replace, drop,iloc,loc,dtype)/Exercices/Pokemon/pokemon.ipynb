```python
# Import the Pandas library
import pandas as pd
```

```python
# File to load
# Create a reference the CSV file desired
df = pd.read_csv('pokemon.csv')
# Read the CSV into a Pandas DataFrame

# Print the first five rows of data to the screen
df.head(5)
```

```python
# Read with Pandas
```

```python
# Extract the following columns: "Type 1", "HP", "Attack", "Sp. Atk", "Sp. Def", and "Speed"
new_df = df[['Type 1','HP', 'Attack', 'Sp. Atk', 'Sp. Def', 'Speed']]
new_df.head(5)
```

```python
# Create a dataframe of the average stats for each type of pokemon.
type_df = new_df.groupby(['Type 1']).mean()
type_df.head(5)
```

```python
# Calculate the total power level of each type of pokemon by adding all of the stats together.
# Place the results into a new column.
new_df['Total Power'] = new_df['Attack'] + new_df['Sp. Atk']+ new_df['Sp. Def']+ new_df['Speed']
new_df
```

```python
# Bonus: Sort the table by strongest type and export the resulting table to a new CSV.
strongest_pokemon = new_df.sort_values(["Total Power"], ascending=False)
strongest_pokemon.reset_index(inplace=True)
strongest_pokemon
```

```python
new_df.to_excel("pokemon_rankings.xlsx", index=False)
```
