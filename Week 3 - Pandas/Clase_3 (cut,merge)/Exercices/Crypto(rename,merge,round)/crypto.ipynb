```python
# Import Dependencies
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
bitcoin_csv = "bitcoin_cash_price.csv"
dash_csv = "dash_price.csv"
```

```python
bitcoin_df = pd.read_csv(bitcoin_csv)
dash_df = pd.read_csv(dash_csv)
```

```python
bitcoin_df.head()
```

```python
dash_df.head()
```

```python
# Merge the two DataFrames together based on the Dates they share
merged_t = pd.merge(bitcoin_df, dash_df, on='Date')
merged_t.head()
```

```python
# Rename columns so that they are differentiated
renamed_merged_t = merged_t.rename(columns={"Open_x": "Open_bitcoin", "High_x": "High_Bitcoin", 
                         "Low_x":"Low_Bitcoin", "Close_x":"Close_Bitcoin", 
                         "Volume_x":"Volume_Bitcoin", "Market Cap_x": "Market_Cap_Bitcoin", 
                         "Open_y":"Open_Dash", "High_y":"High_Dash", "Low_y":"Low_Dash","Close_y":"Close_Dash",
                         "Volume_y":"Volume_Dash", "Market Cap_y":"Market_Cap_Dash"})
renamed_merged_t.head(5)
```

```python
easy_renamed_merged_t = pd.merge(bitcoin_df, dash_df, on="Date", suffixes=('_bitcoin', '_dash'))
easy_renamed_merged_t.head()
```

# alternatively you can set your suffixes when the merge occurs

```python
# Collecting best open for Bitcoin and Dash
b_open_bitcoin = easy_renamed_merged_t['Open_bitcoin'].max()
b_open_dash = easy_renamed_merged_t['Open_dash'].max()
b_open_bitcoin
b_open_dash

# Collecting best close for Bitcoin and Dash
b_close_bitcoin = easy_renamed_merged_t['Close_bitcoin'].max()
b_close_dash = easy_renamed_merged_t['Close_bitcoin'].max()
b_close_bitcoin
b_close_dash

# Collecting the total volume for Bitcoin and Dash
total_vol_bitcoin = round(easy_renamed_merged_t['Volume_bitcoin'].sum()/1000000,2)
total_vol_dash = round(easy_renamed_merged_t['Volume_dash'].sum()/1000000,2)
total_vol_bitcoin
total_vol_dash
```

```python
# Creating a summary DataFrame using above values
summary_df = pd.DataFrame({"Best Bitcoin Open": [b_open_bitcoin],
                           "Best Bitcoin Close": [b_close_bitcoin],
                           "Total Bitcoin Volume": str(total_vol_bitcoin)+" million",
                           "Best Dash Open": [b_open_dash],
                           "Best Dash Close": [b_close_dash],
                           "Total Dash Volume": str(total_vol_dash)+" million"})
summary_df
```
