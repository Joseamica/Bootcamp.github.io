## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#agrupamos la columna gender
p_m = df.groupby('Gender')

#guardamos en una variable el df de p_m, tiramos las columnas sobrantes, y agregamos las nuevas
gender_demo = p_m.count()
gender_demo = gender_demo.drop(['Purchase ID',"Age",'SN','Item ID'], axis=1)
gender_demo.columns = ['Total Count', 'Percentage Players']
#Ordenamos en orden
sorted_gender_demo = gender_demo.sort_values(by='Total Count', ascending=False)

#en la variable percentage dividimos los valores de cada genero / el total de las compras * 100
percentage = (sorted_gender_demo['Total Count']/sorted_gender_demo['Total Count'].sum()) * 100
#Asignamos en la columna Percentage Players los valores calculados arriba y damos formato
sorted_gender_demo['Percentage Players'] = percentage.map("{:.2f}".format)


sorted_gender_demo

```


## Purchasing Analysis (Gender)

```python
#agrupamos Gender
g_gender = df1.groupby('Gender')

#asignamos en una variable los valores de g_gender.
purchasing_an = g_gender.count()
#Tiramos las tablas que no nos sirven.
purchasing_an = purchasing_an.drop(['SN','Purchase ID'], axis=1)
#Renombramos las columnas
purchasing_an.columns = ['Purchase Count', 'Average Purchase Price', 'Total Purchase Value', 'Avg Total Purchase per Person']


#definimos una variable en la cual saca el total del monto comprado. seleccionamos la columna Purchase Count y sumamos
purchase_count_col = purchasing_an['Purchase Count'].sum()

# sumamos todos los valores en 'Price' por genero
total_purchase_price_gender = (g_gender.sum())['Price']

#Asignamos en las columnas las calculaciones para llegar al resultado
purchasing_an['Total Purchase Value'] = total_purchase_price_gender.map("${:.2f}".format)
purchasing_an['Average Purchase Price'] = (total_purchase_price_gender / purchasing_an['Purchase Count']).map("${:.2f}".format)
purchasing_an['Avg Total Purchase per Person'] = (total_purchase_price_gender / sorted_gender_demo['Total Count']).map("${:.2f}".format)

purchasing_an
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 10, 15, 20, 25, 30, 35, 40,100]
df.head()
```

```python

#bining
bins = [0, 10, 15, 20, 25, 30, 35, 40,100]
group_names = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
df["Age Categories"] = pd.cut(df.Age, bins, labels=group_names, right=False)
#get age categories grouped
grouped_percentage_players = df.groupby(['Age Categories'])
#get price and count, then add title
grouped_price = pd.DataFrame(grouped_percentage_players['Price'].count())
grouped_price.columns = ['Total Count']
#get grouped go to age and count
grouped_ages = grouped_percentage_players['Age'].count() 
grouped_purchases_count = grouped_price['Total Count'].sum()

percentage_of_players = grouped_ages / grouped_purchases_count
grouped_price['Percentage of Players'] = (percentage_of_players * 100).map("{:.2f}".format)


grouped_price

```
